```python
import pandas as pd
from sqlalchemy import create_engine, text

DB_ID = "admin"
DB_PW = "admin"
DB_HOST = "172.17.0.2"
DB_PORT = "5432"
DB_NAME = "postgres"

try:
    engine = create_engine(f'postgresql+psycopg2://{DB_ID}:{DB_PW}@{DB_HOST}:{DB_PORT}/{DB_NAME}')
    
    train_df = pd.read_csv('Train.csv')
    test_df = pd.read_csv('Test.csv')
    print("CSV 파일을 성공적으로 읽었습니다.")
    
    train_df.to_sql(name="train", con=engine, if_exists='replace', index=False)
    print("'train' 테이블에 데이터 저장이 완료되었습니다.")
    
    test_df.to_sql(name="test", con=engine, if_exists='replace', index=False)
    print("'test' 테이블에 데이터 저장이 완료되었습니다.")
    
    with engine.connect() as conn:
        conn.execute(text("""CREATE TABLE IF NOT EXISTS predict (
            id INTEGER,
            loadtime TIMESTAMP,
            predict INTEGER
        )"""))
        conn.commit()
    print("'predict' 빈 테이블이 생성되었습니다.")

except Exception as e:
    print(f"오류 발생: {e}")
finally:
    engine.dispose()
    print("DB 연결이 종료되었습니다.")
```
